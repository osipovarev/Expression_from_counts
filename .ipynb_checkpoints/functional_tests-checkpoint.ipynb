```python
%matplotlib inline
import numpy as np
from glob import glob
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib
# from scipy.stats import mannwhitneyu
# from scipy.stats import f_oneway
# from scipy.stats import ttest_ind 
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
dir_path = '/Users/osipova/Documents/LabDocs/NectarivoryProject/Manuscript_Nectar_Genomics/mlxipl_functional_assays_Konstantin/'
file_name = 'summary_functional_assays_mlxipl.tsv'
file_data = glob(dir_path + file_name)[0]
df = pd.read_csv(file_data, sep='\t')
df.head()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(10, 10))
p = sns.color_palette('pastel', n_colors=4)
p = [p[3], p[0]]
sns.set_palette(p)

assays = list(set(df['assay']))
N = len(assays)

for i in range(N):
    j = i // 3
    k = i % 3
    axi = axes[j][k]
    assay = assays[i]
    
    sns.barplot(ax=axi, data=df[(df['assay'] == assay) & (df['glucose'] == 25)],
                x='sample', y='value', errwidth=1.2, capsize=.1, ci='sd')
    sns.stripplot(ax=axi, data=df[(df['assay'] == assay) & (df['glucose'] == 25)],
                  x='sample', y='value', color='black', s=7)
    axi.set_xticklabels(axi.get_xticklabels(), rotation=40, ha="right")
    axi.set_title(assay)
    axi.set_xlabel('')
    axi.set_ylabel('')
    
plt.tight_layout()
sns.despine(fig=None, ax=None, left=False, bottom=False, offset=None, trim=False)
```
